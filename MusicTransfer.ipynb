```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

SCOPES = "playlist-read-private playlist-read-collaborative user-read-private"

auth = SpotifyOAuth(
    scope=SCOPES,
    open_browser=False,
    cache_path=".spotify_token_cache"
)

auth_url = auth.get_authorize_url()
print("Open this URL in a NEW TAB and approve:\n")
print(auth_url)

redirected_url = input("\nAfter approving, paste the FULL redirect URL:\n").strip()
code = auth.parse_response_code(redirected_url)
token_info = auth.get_access_token(code, as_dict=True)

sp = spotipy.Spotify(auth=token_info["access_token"])
me = sp.current_user()
print("✅ Logged in as:", me.get("display_name"), "|", me.get("id"))
print("Country:", me.get("country"))
```

```python
playlists = sp.current_user_playlists(limit=50)
list_of_playlists = []
for i, p in enumerate(playlists["items"], 1):
    name = p.get("name", "Unknown")
    total = (p.get("items") or {}).get("total", "Unknown")
    print(f"{i}. {name} | Tracks: {total} | id: {p.get('id')}")
    list_of_playlists.append(p.get('id'))
print(list_of_playlists)
```

```python
PLAYLIST_IDS = [
    '2Igmh2COF8IG3WysVNCICd'
    '6b0pnbjp3esBS1iT7GK5u5',
    '3iDRpFn39vjEPo5Ci87C1k',
    '0aAN01ZokevoY0GKwKMJzj',
    '6apnZ64aa4dkpgBUTKKhng',
    '3x0w6VzHlDCYZWr4M3Abzp',
    '5b2iOHrUSGZ7tXLPExNcUn'
]

for pid in PLAYLIST_IDS:

    playlist_info = sp.playlist(pid)
    playlist_name = playlist_info["name"]

    print("\n==============================")
    print("Playlist:", playlist_name)
    print("==============================\n")

    items = sp.playlist_items(pid, limit=100)

    shown = 0

    for it in items["items"]:
        track = it.get("track") or {}

        if track.get("type") != "track":
            continue

        name = track.get("name")
        artists = ", ".join([a["name"] for a in track.get("artists", [])])

        print("-", name, "—", artists)

        shown += 1

        if shown >= 5:
            break
```

```python
def readable_count(pid, market="IN"):
    resp = sp.playlist_items(
        pid,
        limit=100,
        market=market,
        fields="items(track(name,artists(name)),is_local),total"
    )
    total = resp.get("total")
    none_ct = sum(1 for it in resp["items"] if it.get("track") is None)
    ok_ct = sum(1 for it in resp["items"] if it.get("track") is not None)
    return total, ok_ct, none_ct

for pid in list_of_playlists:
    p = sp.playlist(pid)
    total, ok_ct, none_ct = readable_count(pid, market="IN")
    print(f"{p['name']}: total={total}, readable_tracks={ok_ct}, null_tracks={none_ct}")
```
