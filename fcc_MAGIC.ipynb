# INDEXING

```python
# Way 1- Setting index during reading file.
df3 = pd.read_csv('world_population.csv', index_col='Country')
display(df3.head())

# Reversing Indexing
df3.reset_index(inplace=True)
df3.head()

# Way 2- Setting after reading the file (NOTE: inplace = True is mandatory to permanently make the desired column as index)
df3.set_index('Country', inplace=True)
df3.head()
```

# Searching through Indexed Column

```python
df3.loc['India']
# df3.loc['ASM'] ## Will throw an error because the value "ASM" is from the CCA3 column and the CCA3 column in not the indexed column.

df3.iloc[3]
```

# Multi Indexing

```python
df3.reset_index(inplace=True)
df3.set_index(['Continent', 'Country'], inplace=True)
df3.head()

display(df3.sort_index())
```

# Searching through multi level indexing

```python
display(df3.loc['Africa'])
# df3.loc['Angola'] # Will throw an error as we're looking at the second level indexing.
df3.loc['Africa', 'Angola'] # Searching through multi level indexing
```

# GROUP BY AND AGGREGATING

```python
df4 = pd.read_csv('world_population.csv')
df_continent_grp = df4.groupby('Continent')
df_continent_grp.mean(numeric_only=True).sort_values(by = 'Rank', ascending=False).filter(items=['Continent', 'Rank', '2022 Population'])

# Usually the str columns are ignored when you're grouping and aggregating, but if error persists, try using " numeric_only = True".
```

```python
df_continent_grp.count().sort_values(by = 'Rank', ascending= False).filter(['Country'])
```

# Aggregating a specific column and multiple aggregations and describe() function.

```python
df_agg_grp = df4.groupby('Continent').agg({'2022 Population': ['sum', 'mean'], 'Rank': 'max'})
display(df_agg_grp)

df_desc = df4.groupby('Continent').describe()
df_desc
```

# idxmax and idxmin to get desired data results of grouped data

```python
display(df4.groupby('Continent')['2022 Population'])

# Group by 'Continent' and find the index of the country with the minimum '2022 Population'
idx_least_pop = df4.groupby('Continent')['2022 Population'].idxmin()
idx_most_pop = df4.groupby('Continent')['2022 Population'].idxmax()
print(idx_least_pop)

# Use these indices to get the full rows from the original DataFrame and assign to hi_pop_con
# We can use both loc and iloc here, but if the index is an non numeric value like country, the use only LOC.
low_pop_con = df4.iloc[idx_least_pop]
high_pop_con = df4.loc[idx_most_pop]

# Display the desired columns from hi_pop_con
display(low_pop_con[['Continent', 'Country', '2022 Population']])
display(high_pop_con[['Continent', 'Country', '2022 Population']])
```

# Merge Joining and Concatinating

```python
lotr_df = pd.read_csv('LOTR.csv')
lotr2_df = pd.read_csv('LOTR 2.csv')
display(lotr_df)
display(lotr2_df)

# Merging
display(lotr_df.merge(lotr2_df, on =['FellowshipID', 'FirstName'] , how = 'right')) # inner, outer, left, right

# Joining
display(lotr_df.join(lotr2_df, on ='FellowshipID' , how = 'outer', lsuffix='-l', rsuffix='-r'))

# Concatinating
pd.concat([lotr_df, lotr2_df])
```
